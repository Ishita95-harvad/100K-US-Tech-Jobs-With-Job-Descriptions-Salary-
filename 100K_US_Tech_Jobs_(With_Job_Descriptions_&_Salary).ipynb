## Data preparation

### Subtask:
Combine the five dataframes into a single dataframe, add a column indicating the job category, and perform initial data cleaning.


**Reasoning**:
Combine the five dataframes, add a job category column, and perform initial data cleaning as instructed.



```python
import pandas as pd

def clean_and_combine_dataframes(dfs, categories):
    combined_df = pd.concat(dfs, keys=categories, names=['job_category'])
    combined_df = combined_df.reset_index(level=0)
    # Handle missing values for specified columns
    for col in ['company_industry', 'company_revenue', 'ceo_name', 'ceo_photo_url']:
        combined_df[col] = combined_df[col].fillna('Not Available')

    #Initial outlier detection and handling for 'mean_salary'
    #This is a very basic approach. More sophisticated methods can be used later.
    combined_df['mean_salary'] = pd.to_numeric(combined_df['mean_salary'], errors='coerce')
    q1 = combined_df['mean_salary'].quantile(0.25)
    q3 = combined_df['mean_salary'].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    combined_df['mean_salary'] = combined_df['mean_salary'].clip(lower=lower_bound, upper=upper_bound)
    combined_df['mean_salary'].fillna(combined_df['mean_salary'].median(), inplace=True)

    # Convert date columns to datetime objects
    combined_df['date_posted'] = pd.to_datetime(combined_df['date_posted'], errors='coerce')

    # Drop unnamed columns
    unnamed_cols = [col for col in combined_df.columns if 'Unnamed' in col]
    combined_df = combined_df.drop(columns=unnamed_cols)

    return combined_df

dfs = [df_all_jobs, df_cs_jobs, df_ds_jobs, df_it_jobs, df_pm_jobs]
categories = ['All Jobs', 'CS', 'DS', 'IT', 'PM']
combined_df = clean_and_combine_dataframes(dfs, categories)

display(combined_df.head())
display(combined_df.info())
```

## Data analysis

### Subtask:
Perform exploratory data analysis on the combined dataset to gain insights into job market trends.


**Reasoning**:
Analyze salary distributions across different job categories. Calculate and compare the mean, median, and standard deviation of salaries for each category. Visualize these distributions. Investigate the distribution of job titles within each category. Identify the most frequent job titles and their counts. Visualize the top N job titles for each category.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Salary distribution across job categories
plt.figure(figsize=(12, 6))
sns.boxplot(x='job_category', y='mean_salary', data=combined_df)
plt.title('Salary Distribution by Job Category')
plt.xlabel('Job Category')
plt.ylabel('Mean Salary')
plt.show()

salary_stats = combined_df.groupby('job_category')['mean_salary'].agg(['mean', 'median', 'std'])
display(salary_stats)

# Top N job titles in each category
top_n = 10
for category in combined_df['job_category'].unique():
    top_titles = combined_df[combined_df['job_category'] == category]['title'].value_counts().nlargest(top_n)
    print(f"\nTop {top_n} Job Titles in {category}:")
    display(top_titles)
    plt.figure(figsize=(10, 6))
    top_titles.plot(kind='bar')
    plt.title(f'Top {top_n} Job Titles in {category}')
    plt.xlabel('Job Title')
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

**Reasoning**:
Explore geographical trends in job postings. Determine the most popular job locations across all categories and within each category. Analyze the relationship between location and salary. Visualize job distributions on a map.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Most popular job locations
top_locations = combined_df['location'].value_counts().nlargest(10)
print("Top 10 Job Locations:")
display(top_locations)

plt.figure(figsize=(10, 6))
top_locations.plot(kind='bar')
plt.title('Top 10 Job Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Location and salary relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mean_salary', y='location', data=combined_df.sort_values('mean_salary', ascending=False).head(100))
plt.title("Relationship between Location and Salary (Top 100 Highest Salaries)")
plt.xlabel('Mean Salary')
plt.ylabel('Location')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Location trends within each category
for category in combined_df['job_category'].unique():
    top_locations_category = combined_df[combined_df['job_category'] == category]['location'].value_counts().nlargest(5)
    print(f"\nTop 5 Locations in {category}:")
    display(top_locations_category)
    plt.figure(figsize=(8, 4))
    top_locations_category.plot(kind='bar')
    plt.title(f'Top 5 Job Locations in {category}')
    plt.xlabel('Location')
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
